```python
import kagglehub
from kagglehub import KaggleDatasetAdapter

# Load the latest version by specifying the target file name
df = kagglehub.load_dataset(
  KaggleDatasetAdapter.PANDAS,
  "yasserh/titanic-dataset",
  "Titanic-Dataset.csv"
)

print(df.head())
```

```python
import pandas as pd
import numpy as np
```

```python
for i in df.columns:
    if df[i].isna().any():
        print(i)
```

Fixing the missing values in age

```python
df['Age'] = df.groupby('Pclass')['Age'].transform(lambda x: x.fillna(x.median()))
# using the method of grouped median
```

```python
for i in df.columns:
    if df[i].isna().any():
        print(i)
```

Fixing the missing values in Embarked

```python
# Find the most common port
most_common_port = df['Embarked'].mode()[0]

# Fill the missing values
df['Embarked'] = df['Embarked'].fillna(most_common_port)
```

```python
for i in df.columns:
    if df[i].isna().any():
        print(i)
```

Fixing the missing values in Cabin

```python
# Replace missing cabins with 'Unknown'
df['Cabin'] = df['Cabin'].fillna('Unknown')
```

```python
for i in df.columns:
    if df[i].isna().any():
        print(i)
```

Handling Duplicate Values

```python
duplicate_count = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_count}")
```

Converting Data Types

```python
# Convert Sex to 0 and 1
df['Sex'] = df['Sex'].map({'male': 1, 'female': 0})
```

```python
df['Sex'].dtypes
```

Using One hot coding concept for Embarked Column

```python
# Convert Embarked into multiple dummy columns (Embarked_Q, Embarked_S, etc.)
df = pd.get_dummies(df, columns=['Embarked'], drop_first=True)
```

```python
df.columns
```

```python
df['PassengerId'].dtype
```

 Convert PassengerId to string text To avoid Confusion

```python
# Convert PassengerId to string text
df['PassengerId'] = df['PassengerId'].astype(str)
```

```python
# Convert Pclass to a categorical type
df['Pclass'] = df['Pclass'].astype('category')
```

```python
df['Age'] = df['Age'].astype(int)
```

```python
print(df.dtypes)
```

Enhancing the data for analysis

```python
# Rename columns for clarity and consistency
df = df.rename(columns={
    'Pclass': 'Passenger_Class',
    'SibSp': 'Siblings_Spouses',
    'Parch': 'Parents_Children',
    'Embarked_Q': 'Embarked_Queenstown',
    'Embarked_S': 'Embarked_Southampton'
})

#  Force all column names to lowercase for easier typing
df.columns = df.columns.str.lower()
```

```python
# Droping columns that won't be used in mathematical analysis
columns_to_drop = ['passengerid', 'name', 'ticket', 'cabin']
df = df.drop(columns=columns_to_drop, errors='ignore')
```

```python
# Combine columns to create 'family_size'
df['family_size'] = df['siblings_spouses'] + df['parents_children'] + 1  # (+1 includes the passenger themselves)
```
